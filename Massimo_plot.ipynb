```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Set up the paths
cwd = Path.cwd() # current directory; this is where this Python code is located
print(cwd)  
np_data = cwd / 'examples'  # where the Numpy data is saved
print(np_data)
```

```python
#sky = np.load(np_data / 'sky_data.npy')
sky_reduced = np.load(np_data / 'sky_reduced.npy')
htsi_data = np.load(np_data / 'htsi_data.npy')
```

```python
np.shape(sky_reduced)
```

```python
np.shape(htsi_data)
```

```python
# Display original and Hadamard transformed images
fig, axes = plt.subplots(1, 2, figsize=(5, 5))
 
# Modify i_wavelength to probe different wavelengths
i_wavelength = 50  # Change this to probe different wavelengths between 400nm (0) and 800nm (199)
 
# Display original image
#axes[0].imshow(sky[256-64:256+63, 256-64:256+63, i_wavelength], cmap='gray')
axes[0].imshow(sky_reduced[:, :, i_wavelength], cmap='gray')
axes[0].set_title("Original")

 
# Display Hadamard transformed image
axes[1].imshow(htsi_data[:, :, i_wavelength], cmap='gray')
axes[1].set_title("Hadamard")
 
plt.tight_layout()
plt.show()
```
